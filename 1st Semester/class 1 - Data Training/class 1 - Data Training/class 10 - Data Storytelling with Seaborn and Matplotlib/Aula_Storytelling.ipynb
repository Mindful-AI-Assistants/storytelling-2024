```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregar o dataset do Titanic
titanic = sns.load_dataset('titanic')
```

```python
titanic[['sex', 'survived']].value_counts()
```

### Gráfico de Barras - Sobrevivência por Sexo

```python
# Gráfico de Barras - Sobrevivência por Sexo
plt.figure(figsize=(10, 5))

sns.countplot(data=titanic, x='sex', hue='survived')
plt.title('Sobrevivência por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.savefig('fig1.png', dpi=300)
plt.show()
```

### Gráfico de Barras - Sobrevivência por Classe

```python
# Gráfico de Barras - Sobrevivência por Sexo
plt.figure(figsize=(10, 5))

sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Sobrevivência por Classe')
plt.xlabel('Classe')
plt.ylabel('Contagem')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.savefig('fig2.png', dpi=300)
plt.show()
```

### Gráfico de Barras - Sobrevivência por Tamanho da Família

O valor para 'family_size' é calculado somando as colunas 'sibsp' e 'parch', e então adicionando 1.

- 'sibsp' é a coluna que representa o número de irmãos/cônjuges a bordo para cada passageiro.
- 'parch' é a coluna que representa o número de pais/filhos a bordo para cada passageiro.

Portanto, 'family_size' é uma representação do tamanho total da família de cada passageiro a bordo (incluindo o próprio passageiro, daí o '+ 1').

```python
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1

plt.figure(figsize=(10, 5))
sns.countplot(data=titanic, x='family_size', hue='survived')

plt.title('Sobrevivência por Tamanho da Família')
plt.xlabel('Tamanho da Família')
plt.ylabel('Contagem')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.savefig('fig4.png', dpi=300)
plt.show()
```
