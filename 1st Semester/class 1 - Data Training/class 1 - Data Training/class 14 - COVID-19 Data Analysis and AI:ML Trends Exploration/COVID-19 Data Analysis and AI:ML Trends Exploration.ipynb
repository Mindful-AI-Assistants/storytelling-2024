```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

Extraia os dados baixados de https://covid.saude.gov.br/

```python
!unzip /content/HIST_PAINEL_COVIDBR_14jun2024.zip
```

Criar uma lista de `Dataframes`. Estamos aqui fazendo um casting para tornar a coluna de datas em um formato específico de datas do Python ao invés de texto. Também se colocou a coluna de data como o índice.

```python
lista_df = []

anos = (2020, 2021, 2022, 2023, 2024)

for ano in anos:
    for parte in (1,2):
      try:
        arq = f'HIST_PAINEL_COVIDBR_{ano}_Parte{parte}_14jun2024.csv'
        print(f'lendo o arquivo {arq}...')
        lista_df.append( pd.read_csv(arq, sep=';', parse_dates=['data'], index_col=['data']) )
      except:
        # Como não existe 2024_Parte2 ainda tive que colocar isso
        pass
```

Faça aqui o concatenamento das linhas dos `dataframes` da lista em um `dataframe` único.

```python
df = pd.concat(lista_df, axis=0)
df
```

Nosso objetivo é mostrar somente os valores a nível de Brasil. Faça um filtro das linhas baseado na coluna `regiao` somente para a correspondência `Brasil`. Salve este filtro em um `Dataframe`.

```python
df = df[ df['regiao'] == 'Brasil' ]
df
```

```python
df.columns
```

```python
filtro = ['casosAcumulado', 'casosNovos', 'obitosAcumulado', 'obitosNovos']

df = df[filtro]
```

```python
df.tail(50)
```

```python
ax = df['casosAcumulado'].plot(figsize=(15,6), grid=True, title='Acumulado de Casos ao longo dos dias' )
ax.set_xlabel('Dia')
ax.set_ylabel('Casos')
plt.show()
```

```python
ax = df['obitosAcumulado'].plot(figsize=(15,6), grid=True, title='Acumulado de Óbitos ao longo dos dias' )
ax.set_xlabel('Dia')
ax.set_ylabel('Óbitos')
plt.show()
```

```python
px.line(df, y=['casosAcumulado', 'obitosAcumulado'])
```

### Comparar as políticas de combate à COVID-19 em dois estados diferentes

```python
df = pd.concat(lista_df, axis=0)
df
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Filtra municípios
df_munic = df[df.municipio.isin(['Rio de Janeiro', 'São Paulo'])]
```

```python
# Média móvel óbitos novos
df_munic['mva_obitosNovos'] = df_munic['obitosNovos'].rolling(window=7).mean()

# Média móvel casos novos
df_munic['mva_casosNovos'] = df_munic['casosNovos'].rolling(window=7).mean()
```

```python
fig3 = px.line(df_munic.reset_index(),
               x='data',
               y='mva_casosNovos',
               color='municipio',
               title='<b>Média Móvel (7d) Casos Novos</b>'
              )
fig3.show()
```

```python
fig4 = px.line(df_munic.reset_index(),
               x='data',
               y='mva_obitosNovos',
               color='municipio',
               title='<b>Média Móvel (7d) Óbitos Novos</b>')
fig4.show()
```
