```python
# Exemplo do pr√≥prop matplotlib
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie",
    "Chinstrap",
    "Gentoo",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

## Gr√°fico de Pizza

```python
import matplotlib.pyplot as plt

# Carregando os dados
df = sns.load_dataset('mpg')

# Definindo o tamanho da figura
plt.figure(figsize=(10, 6))

# Criando um gr√°fico de linhas para cada origem
for origin in df['origin'].unique():
    subset = df[df['origin'] == origin]
    plt.plot(subset['model_year'], subset['mpg'], label=origin)

# Adicionando um t√≠tulo e uma legenda
plt.title('MPG ao longo dos anos por origem')
plt.legend()

# Mostrando o gr√°fico
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='model_year', y='mpg', hue='origin', data=df)
plt.title('MPG ao longo dos anos por origem')
plt.show()
```

```python
# Variante do gr√°fico de barras
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='mpg', hue='cylinders', data=df)
plt.title('Distribui√ß√£o de MPG por origem e cilindros')
plt.show()
```
