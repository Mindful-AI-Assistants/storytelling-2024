### Consumo de Açúcar e Educação

TABLE 3. Crude prevalence* of sugar-sweetened beverage† consumption ≥1 time/day among adults, by employment status, education, and state — Behavioral Risk Factor Surveillance System, 23 states and District of Columbia, 2013

https://www.cdc.gov/mmwr/volumes/65/wr/mm6507a1.htm

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/noahgift/sugar/master/data/education_sugar_cdc_2003.csv")
df.set_index("State", inplace=True)
```

```python
df
```

```python
for col in df.columns:
  # Remever um padrão com base em regra (regex)
    df[col] = df[col].str.replace(r"\(.*\)", "", regex=True)
    df[col] = pd.to_numeric(df[col])
```

```python
df
```

```python
df.info()
```

```python
!wget https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/us-states.json
```

```python
# Reset the index
df.reset_index(level=0, inplace=True)

import folium
m = folium.Map(location=[36, -102], zoom_start=3)

folium.Choropleth(
    geo_data="us-states.json",
    name='choropleth',
    data=df,
    columns=['State', '<High school'],
    key_on='feature.properties.name',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='<High school Education and Grams Sugar Intake Daily'
).add_to(m)

folium.LayerControl().add_to(m)

m
```

```python
import folium
m = folium.Map(location=[36, -102], zoom_start=3)

folium.Choropleth(
    geo_data="us-states.json",
    name='choropleth',
    data=df,
    columns=['State', 'College graduate'],
    key_on='feature.properties.name',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='College graduate and Grams Sugar Intake Daily'
).add_to(m)

folium.LayerControl().add_to(m)

m
```

```python
df3 = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv')
df3.rename(columns={'state': 'State'}, inplace=True)
df3.set_index('State', inplace=True)
```

```python
df3 = df3.groupby('State').agg('max')
df3.drop(columns=['date','fips'], inplace=True)
df3
```

```python
df4 = df2.merge(df3, how="inner", on="State")
```

```python
df4
```

```python
correlacao = df4.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizar a matriz de correlação com um heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', center=0 )
plt.title('Matriz de Correlação de nossa base de dados')
plt.show()
```
