```python
import matplotlib.pyplot as plt
import numpy as np
```

## Trabalhando com subplots

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

```python
plt.figure()

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
plt.figure()

plt.subplot(1,2,1)
plt.plot(x, np.sin(x))

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
```

```python
plt.figure()

plt.subplot(2,2,1)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.grid()

plt.subplot(2,2,3)
plt.plot(x, np.tan(x))

plt.subplot(2,2,4)
plt.plot(x, np.exp(x))
```

```python
plt.figure()

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

Colocando uma figura com resolução específica

```python
import matplotlib.pyplot as plt
text_kwargs = dict(ha='center', va='center', fontsize=28, color='C1')

px = 1/plt.rcParams['figure.dpi']  # pixel in inches
plt.subplots(figsize=(600*px, 200*px))
plt.text(0.5, 0.5, '600px x 200px', **text_kwargs)
plt.show()
```

Exemplo de dashboard simples

```python
import matplotlib.pyplot as plt

# Como especificar pixels diretamente

px = 1/plt.rcParams['figure.dpi']  # pixel in inches

plt.subplots(figsize=(1920*px, 1080*px))

plt.subplot(2,2,1)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.grid()

plt.subplot(2,2,3)
text_kwargs = dict(ha='center', va='center', fontsize=28, color='C1')
plt.text(0.5, 0.5, 'Minha logo', **text_kwargs)

plt.subplot(2,2,4)
plt.plot(x, np.tan(x))


plt.show()
```

## Mudando cores e estilo


```python
plt.plot(x, np.sin(x-0), color='blue') # cor por extenso
plt.plot(x, np.sin(x-1), color='g') # atalho
plt.plot(x, np.sin(x-2), color='0.75') # Escala de cinza
plt.plot(x, np.sin(x-3), color='#FFDD44') # Código hexadecimal
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x-1, linestyle='-')
plt.plot(x, x-2, linestyle='--')
plt.plot(x, x-3, linestyle='-.')
plt.plot(x, x-4, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

Mudando o intervalo

```python
plt.plot(x, np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5]) # Manual
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

### Trabalahndo com rótulos

```python
plt.plot(x, np.sin(x))
plt.title('Curva do seno')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python

plt.plot(x, np.sin(x), '-g', label= r'sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```
